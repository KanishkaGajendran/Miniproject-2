```python
# After dropping the column
df
```

#**Exploratory Data Analysis(EDA)**

```python
#Top 7 RICE PRODUCTION State Data(Bar_plot)

top_7_states = (df.groupby('State Name')['RICE PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(7).reset_index())
print(top_7_states)
plt.figure(figsize=(5,5))
plt.bar(top_7_states['State Name'],top_7_states['RICE PRODUCTION (1000 tons)'])
plt.title('TOP 7 RICE PRODUCTION state')
plt.xlabel('State Name')
plt.ylabel('RICE PRODUCTION (1000 tons)')
plt.xticks(rotation=90)
plt.show()
```

```python
#Top 5 Wheat Producing States Data
top_5_states=(df.groupby('State Name')['WHEAT PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(5).reset_index())
print(top_5_states)

plt.figure(figsize=(5,5))
plt.bar(top_5_states['State Name'],top_5_states['WHEAT PRODUCTION (1000 tons)'],color='skyblue')
plt.title('TOP 5 WHEAT PRODUCTION state')
plt.xlabel('State Name')
plt.ylabel('WHEAT PRODUCTION (1000 tons)')
plt.xticks(rotation=90)
plt.show()
```

```python
#Top 5 Wheat Producing States percentage(%)(Pie_chart)
top_5_states=(df.groupby('State Name')['WHEAT PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(5).reset_index())
print(top_5_states)
plt.figure(figsize=(5,5))
plt.pie(top_5_states['WHEAT PRODUCTION (1000 tons)'],labels=top_5_states['State Name'],autopct ='%1.1f%%')
plt.title('Top 5 state by wheat production')
plt.show()

```

```python
#TOP 5 OIL SEED PRODUCTION STATE
top_5_states=(df.groupby('State Name')['OILSEEDS PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(5).reset_index())
print(top_5_states)
plt.figure(figsize=(5,5))
plt.bar(top_5_states['State Name'],top_5_states['OILSEEDS PRODUCTION (1000 tons)'],color='orange')
plt.title('TOP 5 OILSEEDS PRODUCTION state')
plt.xlabel('State Name')
plt.ylabel('OILSEEDS PRODUCTION (1000 tons)')
plt.xticks(rotation=90)
plt.show()
```

```python
#TOP 7 SUN FLOWER PRODUCTION BY STATE
top_7_states=(df.groupby('State Name')['SUNFLOWER PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(7).reset_index())
print(top_7_states)
plt.figure(figsize=(5,5))
plt.bar(top_7_states['State Name'],top_7_states['SUNFLOWER PRODUCTION (1000 tons)'],color = 'yellow')
plt.title('TOP 7 SUNFLOWER PRODUCTION state')
plt.xlabel('State Name')
plt.ylabel('SUNFLOWER PRODUCTION (1000 tons)')
plt.xticks(rotation=90)
plt.show()
```

```python
#India's SUGARCANE PRODUCTION From Last 50 Years(Line_plot)

latest_year = df['Year'].max()
df_50 = df[df['Year'] >= latest_year-49]

yearly_sugarcane=(df_50.groupby('Year')['SUGARCANE PRODUCTION (1000 tons)'].sum().reset_index())

plt.figure()
plt.plot(yearly_sugarcane['Year'],yearly_sugarcane['SUGARCANE PRODUCTION (1000 tons)'],marker='o')
plt.xlabel('Year')
plt.ylabel('Sugarcane production')
plt.title("India's SUGARCANE PRODUCTION From Last 50 Years(Line_plot)")
plt.xticks(rotation=90)
plt.show()                 
```

```python
#Rice Production Vs Wheat Production (Last 50y)
last_50_years=sorted(df['Year'].unique())[-50:]
df_50=df[df['Year'].isin(last_50_years)]

prod_50 = df_50.groupby('Year')[['RICE PRODUCTION (1000 tons)','WHEAT PRODUCTION (1000 tons)']].sum()

plt.figure()
plt.plot(prod_50.index,prod_50['RICE PRODUCTION (1000 tons)'],label='Rice')
plt.plot(prod_50.index,prod_50['WHEAT PRODUCTION (1000 tons)'],label='Wheat')
plt.xlabel('Year')
plt.ylabel('Rice vs Wheat production')
plt.title('Rice Production Vs Wheat Production (Last 50y)')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
#Rice Production By West Bengal Districts
WB=df[df['State Name']=='West Bengal']

plt.figure(figsize = (8,6))
plt.bar(WB['Dist Name'],WB['RICE PRODUCTION (1000 tons)'], color='skyblue')
plt.title('Rice Production By West Bengal Districts')
plt.xlabel('Dist Name')
plt.ylabel('RICE PRODUCTION (1000 tons)')
plt.xticks(rotation = 90)
plt.show()
```

```python
#Top 10 Wheat Production Years From UP
UP=df[df['State Name']=='Uttar Pradesh']
Year_wise = UP.groupby('Year')['WHEAT PRODUCTION (1000 tons)'].sum().reset_index()
Top10_years = Year_wise.sort_values(by='WHEAT PRODUCTION (1000 tons)',ascending=False).head(10)
print(Top10_years)

plt.figure(figsize = (8,6))
plt.bar(Top10_years['Year'].astype(str),Top10_years['WHEAT PRODUCTION (1000 tons)'], color='brown')
plt.title('Top 10 Wheat Production Years From UP')
plt.xlabel('Year')
plt.ylabel('WHEAT PRODUCTION (1000 tons)')
plt.xticks(rotation = 90)
plt.show()

```

```python
#Millet Production (Last 50y)

millet_prod = df_50.groupby('Year')[['PEARL MILLET PRODUCTION (1000 tons)','FINGER MILLET PRODUCTION (1000 tons)']].sum()
plt.figure()
plt.plot(millet_prod.index,millet_prod['PEARL MILLET PRODUCTION (1000 tons)'],label ='Pearl millet')
plt.plot(millet_prod.index,millet_prod['FINGER MILLET PRODUCTION (1000 tons)'],label ='Finger millet')
plt.xlabel('Year')
plt.ylabel('Production')
plt.title('Millet Production (Last 50y)')
plt.legend()
plt.xticks(rotation =90)
plt.show()
```

```python
#Sorghum Production (Kharif and Rabi) by Region
plt.figure(figsize=(10,6))
plt.scatter(df['KHARIF SORGHUM AREA (1000 ha)'],df['KHARIF SORGHUM PRODUCTION (1000 tons)'],color='skyblue',label='Kharif',alpha=0.8)
plt.scatter(df['RABI SORGHUM AREA (1000 ha)'],df['RABI SORGHUM PRODUCTION (1000 tons)'],color='blue',label='Rabi',alpha=0.8)
plt.title('Sorghum Production (Kharif and Rabi) by Region')
plt.xlabel('AREA (1000 ha)')
plt.ylabel('Production (1000 tons)')
plt.legend()
plt.show()
```

```python
#Top 7 States for Groundnut Production
top_7_states=(df.groupby('State Name')['GROUNDNUT PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(7).reset_index())
print(top_7_states)

plt.figure(figsize=(5,5))
plt.pie(top_7_states['GROUNDNUT PRODUCTION (1000 tons)'],labels=top_7_states['State Name'],autopct='%1.1f%%')
plt.title('Top 7 States for Groundnut Production')
plt.show()
```

```python
#Soybean Production by Top 5 States and Yield Efficiency
top_5_states=(df.groupby('State Name')['SOYABEAN PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(5))
print(top_5_states)

yield_efficiency = (df[df['State Name'].isin(top_5_states.index)].groupby('State Name')['SOYABEAN YIELD (Kg per ha)'].mean())

fig,ax1=plt.subplots()

ax1.bar(top_5_states.index,top_5_states.values)
ax1.set_xlabel('State Name')
ax1.set_ylabel('Soyabean Production')
ax1.set_title('Soybean Production by Top 5 States and Yield Efficiency')
ax1.tick_params(axis='x',rotation=30)

ax2=ax1.twinx()
ax2.plot(yield_efficiency.index,yield_efficiency.values,marker='o')
ax2.set_ylabel('Yield Efficiency')
fig.tight_layout()
plt.show()

               
```

```python
#Oilseed Production in Major States
oilseed_prod=df.groupby('State Name')['OILSEEDS PRODUCTION (1000 tons)'].sum().reset_index()
maj_states=oilseed_prod.sort_values(by='OILSEEDS PRODUCTION (1000 tons)',ascending=False).head(5)

plt.figure(figsize=(10,6))
sns.barplot(x='State Name',y='OILSEEDS PRODUCTION (1000 tons)',data=maj_states)
plt.title('Oilseed Production in Major States')
plt.xlabel('State')
plt.ylabel('OIL SEEDS PRODUCTION')
plt.show()
```

```python
#Impact of Area Cultivated on Production (Rice, Wheat, Maize)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Rice
axes[0].scatter(df['RICE AREA (1000 ha)'], df['RICE PRODUCTION (1000 tons)'], color='red')
axes[0].set_title('Rice: Area vs Production')
axes[0].set_xlabel('Area')
axes[0].set_ylabel('Production')

# Wheat
axes[1].scatter(df['WHEAT AREA (1000 ha)'], df['WHEAT PRODUCTION (1000 tons)'], color='yellow')
axes[1].set_title('Wheat: Area vs Production')
axes[1].set_xlabel('Area')
axes[1].set_ylabel('Production')

# Maize
axes[2].scatter(df['MAIZE AREA (1000 ha)'], df['MAIZE PRODUCTION (1000 tons)'], color='green')
axes[2].set_title('Maize: Area vs Production')
axes[2].set_xlabel('Area')
axes[2].set_ylabel('Production')

plt.show()
```

```python
#Rice vs. Wheat Yield Across States
plt.figure(figsize=(10,6))
plt.scatter(df['State Name'],df['RICE YIELD (Kg per ha)'],color='red',label='Rice',marker='x',alpha=0.8)
plt.scatter(df['State Name'],df['WHEAT YIELD (Kg per ha)'],color='orange',label='Wheat',marker='o',alpha=0.8)
plt.title('Rice vs. Wheat Yield Across States')
plt.xlabel('State')
plt.ylabel('Yield (Kg per ha)')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

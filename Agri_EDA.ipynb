#**Exploratory Data Analysis(EDA)**

```python
#Top 7 RICE PRODUCTION State Data(Bar_plot)

rice_prod=df[['State Name','RICE PRODUCTION (1000 tons)']].sort_values(by='RICE PRODUCTION (1000 tons)',ascending =False).head(7)
print(rice_prod[['State Name','RICE PRODUCTION (1000 tons)']])

plt.figure(figsize=(5,5))
plt.bar(rice_prod['State Name'],rice_prod['RICE PRODUCTION (1000 tons)'])
plt.title('TOP 7 RICE PRODUCTION state')
plt.xlabel('State Name')
plt.ylabel('RICE PRODUCTION (1000 tons)')
plt.show()
```

```python
#Top 5 Wheat Producing States Data
wheat_prod=df[['State Name','WHEAT PRODUCTION (1000 tons)']].sort_values(by='WHEAT PRODUCTION (1000 tons)',ascending =False).head(5)
print(wheat_prod[['State Name','WHEAT PRODUCTION (1000 tons)']])

plt.figure(figsize=(5,5))
plt.bar(wheat_prod['State Name'],wheat_prod['WHEAT PRODUCTION (1000 tons)'])
plt.title('TOP 5 WHEAT PRODUCTION state')
plt.xlabel('State Name')
plt.ylabel('WHEAT PRODUCTION (1000 tons)')
plt.show()
```

```python
#Top 5 Wheat Producing States percentage(%)(Pie_chart)
wheat_prod=df[['State Name','WHEAT PRODUCTION (1000 tons)']].sort_values(by='WHEAT PRODUCTION (1000 tons)',ascending =False).head(5)

plt.figure(figsize=(5,5))
plt.pie(wheat_prod['WHEAT PRODUCTION (1000 tons)'],labels=wheat_prod['State Name'],autopct ='%1.1f%%')
plt.title('Top 5 state by wheat production')
plt.show()

```

```python
#TOP 5 OIL SEED PRODUCTION STATE
oilseed_prod=df[['State Name','OILSEEDS PRODUCTION (1000 tons)']].sort_values(by='OILSEEDS PRODUCTION (1000 tons)',ascending =False).head(5)
print(oilseed_prod[['State Name','OILSEEDS PRODUCTION (1000 tons)']])

plt.figure(figsize=(5,5))
plt.bar(oilseed_prod['State Name'],oilseed_prod['OILSEEDS PRODUCTION (1000 tons)'])
plt.title('TOP 5 OILSEEDS PRODUCTION state')
plt.xlabel('State Name')
plt.ylabel('OILSEEDS PRODUCTION (1000 tons)')
plt.show()
```

```python
#TOP 7 SUN FLOWER PRODUCTION BY STATE
sunflower_prod=df[['State Name','SUNFLOWER PRODUCTION (1000 tons)']].sort_values(by='SUNFLOWER PRODUCTION (1000 tons)',ascending =False).head(7)
print(sunflower_prod[['State Name','SUNFLOWER PRODUCTION (1000 tons)']])

plt.figure(figsize=(5,5))
plt.bar(sunflower_prod['State Name'],sunflower_prod['SUNFLOWER PRODUCTION (1000 tons)'])
plt.title('TOP 7 SUNFLOWER PRODUCTION state')
plt.xlabel('State Name')
plt.ylabel('SUNFLOWER PRODUCTION (1000 tons)')
plt.show()
```

```python
#India's SUGARCANE PRODUCTION From Last 50 Years(Line_plot)

#to find last 50 years data
latest_year = df['Year'].max()
first_year = latest_year-49
last_50 = df[(df['Year'] >= first_year) & (df['Year'] <= latest_year)]

plt.figure(figsize=(5,5))
plt.plot(last_50['Year'],last_50['SUGARCANE PRODUCTION (1000 tons)'],)
plt.title("India's SUGARCANE PRODUCTION From Last 50 Years")
plt.xlabel('Year')
plt.ylabel('SUGARCANE PRODUCTION (1000 tons)')
plt.grid(linestyle = '--')
plt.show()
```

```python
#Rice Production Vs Wheat Production (Last 50y)
plt.figure(figsize=(10,6))
plt.bar(last_50['RICE PRODUCTION (1000 tons)'],label='RICE',height= 1,width = 0.4,alpha =0.8)
plt.bar(last_50['WHEAT PRODUCTION (1000 tons)'],label='WHEAT',height=1,width = 0.4,alpha =0.8)
plt.title('Rice Production Vs Wheat Production (Last 50y)')
plt.xlabel('Year')
plt.ylabel('Production')
plt.legend()
plt.show()
```

```python
#Rice Production By West Bengal Districts
WB=df[df['State Name']=='West Bengal']

plt.figure(figsize = (8,6))
plt.bar(WB['Dist Name'],WB['RICE PRODUCTION (1000 tons)'], color='skyblue')
plt.title('Rice Production By West Bengal Districts')
plt.xlabel('Dist Name')
plt.ylabel('RICE PRODUCTION (1000 tons)')
plt.xticks(rotation = 90)
plt.show()
```

```python
#Top 10 Wheat Production Years From UP
UP=df[df['State Name']=='Uttar Pradesh']
Year_wise = UP.groupby('Year')['WHEAT PRODUCTION (1000 tons)'].sum().reset_index()
Top10_years = Year_wise.sort_values(by='WHEAT PRODUCTION (1000 tons)',ascending=False).head(10)
print(Top10_years)

plt.figure(figsize = (8,6))
plt.bar(Top10_years['Year'].astype(str),Top10_years['WHEAT PRODUCTION (1000 tons)'], color='brown')
plt.title('Top 10 Wheat Production Years From UP')
plt.xlabel('Year')
plt.ylabel('WHEAT PRODUCTION (1000 tons)')
plt.xticks(rotation = 90)
plt.show()

```
